1.  **Cuáles son las ventas de cada uno de los productos vendidos por categoría y por cada uno de los vendedores, indique aquí los nombres de estado civil sexo y
tipo de identificación de cada vendedor en la consulta.**

```python
''' SELECT P.IdProducto, P.Nombre, P.IdCategoria, DV.Total, CONCAT(VE.Nombre1, VE.Nombre2, VE.Apellido1, VE.Apellido2) as NombreCompleto, VE.EstadoCivil, VE.Sexo, VE.TipoDeIdentificacion
    FROM TblVenta AS V
    INNER JOIN TblDetalleVenta AS DV ON V.IdFactura = DV.IdFactura
    INNER JOIN TblProducto AS P ON DV.IdProducto = P.IdProducto
    LEFT JOIN TblVendedor AS VE ON VE.Identificacion = V.Vendedor;
'''
```

2. **Cuáles son los productos que han tenido mayor venta y a qué vendedor pertenece?**

```python
''' SELECT P.IdProducto, P.Nombre, SUM(DV.Total) AS TotalVendido, CONCAT(VE.Nombre1, VE.Apellido1) AS NombreVendedor, VE.Identificacion
    FROM TblVenta AS V
    INNER JOIN TblDetalleVenta AS DV ON V.IdFactura = DV.IdFactura
    INNER JOIN TblProducto AS P ON DV.IdProducto = P.IdProducto
    LEFT JOIN TblVendedor AS VE ON VE.Identificacion = V.Vendedor
    GROUP BY P.IdProducto, VE.Identificacion
    ORDER BY TotalVendido DESC
    LIMIT 10;
'''
```

4. **¿Cuáles son algunas estrategias que usarías para optimizar el rendimiento de consultas SQL en grandes conjuntos de datos?**

R/.
- Manejar los JOIN dentro de las consultas por medio de las columnas que llevan los ID,
ya que son numericos y unicos, asi se puede asegurar que no hay informacion replicada y que no se está procesando sobre datos alfanumericos, los cuales consumen mas recursos.

- Cuando se hace una consulta establecer periodos de tiempo, para no cargar registros innecesarios.

- Cuando se necesita hacer varias/repetir consultas o consultas complejas, es mejor usar tablas temporales ya que solo viven durante la consulta y por lo tanto permite usarlas repetidas ocasiones durante la consulta.

- Consultar solo las columnas necesarias, usar * cuando sea realmente necesario.

- Cuando hay grandes volumenes de datos, consultar por lotes, es decir, filtrar la informacion a casos/filtros en alguna variable, con el fin de reducir el tamaño de la información solicitada.

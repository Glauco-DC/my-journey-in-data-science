```python
#Replace $ and ',' for '' in price
autos['price'] = autos['price'].str.replace('$','', regex=True)
autos['price'] = autos['price'].str.replace(',','', regex=True)
autos['price'] = autos['price'].astype(int)
```

```python
#Replace km and ',' for '' in odometer
autos['odometer'] = autos['odometer'].str.replace('km','', regex=True)
autos['odometer'] = autos['odometer'].str.replace(',','', regex=True)
autos['odometer'] = autos['odometer'].astype(int)
```

```python
#rename odometer column to odometer_km

autos.rename({'odometer': 'odometer_km'}, axis=1, inplace=True)

autos.head()
```

```python
top_20 = autos['brand'].value_counts(ascending=False).head(20).index
print(top_20)
```

We select the top 20 brands by amount of cars, to explore data

```python
#Mean price per brand
avg_price_brand = {}

for brand in top_20:
    # Make in one line with .loc[mask, index_to_work]
    mean = autos.loc[autos['brand'] == brand, 'price'].mean().round(2)
    #Two line, first mask and results autos filter after mean
    '''sel = autos[autos['brand'] == brand]
    mean = sel['price'].mean()'''
    
    avg_price_brand[brand] = mean
    

print(avg_price_brand)
```

```python
#Milage mean per brand
avg_milage_brand = {}
for brand in top_20:
    # Make in one line with .loc[mask, index_to_work]
    mean = autos.loc[autos['brand'] == brand, 'odometer_km'].mean().round(2)
    #Two line, first mask and results autos filter after mean
    '''sel = autos[autos['brand'] == brand]
    mean = sel['price'].mean()'''
    
    avg_milage_brand[brand] = mean

print(avg_milage_brand)
```

```python
#Convert both dictionaries to series objects, using the series constructor.
#Create a dataframe from dictionary. Dictionary->Series->DataFrame
means = pd.DataFrame(pd.Series(avg_price_brand), columns = ['mean_price'])
#Add a new column Dic->Serie->DataFrame
means['mean_milage'] = pd.DataFrame(pd.Series(avg_milage_brand))

means
```
